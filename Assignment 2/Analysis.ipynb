```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("Data to Analysis.xlsx")
foots=pd.read_excel("All players data link.xlsx")
```

```python
data=pd.concat([data,foots["Foot"]],axis=1)
```

# starting analysing

```python
data
```

```python
data.info()
```

```python
plt.figure(figsize=(16,18))
sns.histplot(data=data, x="Overall Rating", hue="Position", multiple="stack")
```

```python
plt.figure(figsize=(16,18))

sns.displot(data, x="Overall Rating", y="Nation",hue="Foot",height=12)
```

```python
plt.figure(figsize=(12,24))
sns.boxplot(data=data, x="Overall Rating", y="Nation")
```

```python
plt.figure(figsize=(18,12))
sns.catplot(x="Dribble",y="Nation",data=data,height=12)
```

```python
plt.figure(figsize=(18,12))

sns.heatmap(data.corr())
```
